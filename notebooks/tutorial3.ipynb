The file is 2.5 GB - my machine only has 4 GB RAM, and it doesn't fit in memory

````python
gdf = aseg_gdf2.read(path / "AUS_10013_Musgrave_EM", engine="dask")
````

The default engine is ``"pandas"``. You can also change it with an attribute:

```python
gdf.engine = "dask"
```

````python
gdf.engine
````

However, everything works a similar way. You still get the DataFrame in the same manner:

````python
df = gdf.df()
df
````

Have a read of the [Dask documentation for DataFrame](https://docs.dask.org/en/stable/dataframe.html#scope) for some tips on how dask works.

The main tip is that nothing actually gets calculated until you call ``.compute()``.

````python
df.columns[:30]
````

````python
df.Easting.min(), df.Easting.max()
````

Huh that's not much use! Let's try:

````python
easting_min = df.Easting.min().compute()
easting_max = df.Easting.max().compute()
````

````python
easting_min, easting_max
````

Or let's get a full list of flightlines:

````python
lines = df["Line"].unique().compute()
````

And extract just one flightline:

````python
line = df[df.Line == 500301].compute()
````
